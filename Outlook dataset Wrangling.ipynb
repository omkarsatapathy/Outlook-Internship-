```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv(r"C:\Users\kiit\Desktop\OMKAR\Outlook dataset.csv")
df.head()
```

```python
df=df.drop('Your Name',axis=1)
```

```python
df= df.dropna(how='any')
```

```python
df.head()
```

```python
df['quality']=df['quality'].replace({1:'No',2:"Yes"})
```

```python
quality=pd.get_dummies(df['quality'],drop_first=True)
quality.head()
df['color_scale'].value_counts()
df['color_scale']= df['color_scale'].replace({1:'color',2:'Gray Scale'})
```

```python
color_scale= pd.get_dummies(df['color_scale'], drop_first=True)
color_scale.head()
```

```python
df['regular']= df['regular'].replace({1:'Yes',2:'No'})
```

```python
regular=pd.get_dummies(df['regular'], drop_first=True)
regular.head()
```

```python
df['length']=df['length'].replace({1:'Discriptive',2:'short'})
```

```python
length=pd.get_dummies(df['length'], drop_first=True)
length.head()
```

```python
df['kind'] = df['kind'].replace({1:'Economic', 2:'Scientific', 3:"Crime", 4:"Politics", 5:"OTH"})
kind = pd.get_dummies(df['kind'], )
kind.head()
```

```python
df=pd.concat([df,color_scale,regular,length, kind, quality],axis=1)
df.head(10)
```

```python
df.drop(['color_scale','length','regular','kind','quality','What is your Favorite Magazine Brand ?'], axis = 1)
df
```

```python
df[['Price','Ratings']].plot();
```

```python
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x= df.drop(['Price'],axis=1)
y=df[['Price']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.10,random_state=1)
```

```python
m= x_train.pop('color_scale')
c=x_train.pop('length')
b=x_train.pop('regular')
ct=x_train.pop('kind')
n=x_train.pop('quality')
w=x_train.pop('What is your Favorite Magazine Brand ?')
x_train
```

```python
reg_model = LinearRegression()
reg_model.fit(x_train, y_train)
```

```python
for idx, col_name in enumerate(x_train.columns):
 print(' {} is {}'.format(col_name, reg_model.coef_[0][idx]))
```

```python
intercept = reg_model.intercept_[0]
print('the intercept for the model is {}'.format(intercept))
```

```python
reg_model.score(x_train, y_train)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
poly = PolynomialFeatures(degree=2, interaction_only=True)
x_train2 = poly.fit_transform(x_train)
x_test2 = poly.fit_transform(x_test)
poly_clf = linear_model.LinearRegression()
poly_clf.fit(x_train2, y_train)
y_pred = poly_clf.predict(x_test2) 
# print (y_pred)
# in sample (training) R^2 will always improve with the number of variables !
print(poly_clf.score(x_train2, y_train))
```

```python
print(poly_clf.score(x_test2,y_test))
```
